```python
import scrapy

class ArticleSpider(scrapy.Spider):
    name = 'articles'
    # change this to whatever page you want to scrape
    start_urls = ['https://www.autopunditz.com/two-wheeler-sales-figures']

    def parse(self, response):
        for article in response.css('article'):
            link = article.css('a::attr(href)').get()
            title = article.css('a::text').get()
            if link and title:
                yield {
                    'title': title.strip(),
                    'url': response.urljoin(link),
                }
```
