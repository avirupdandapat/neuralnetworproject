```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create the DataFrame utilizing Pandas 
data = pd.read_csv('sales_data_raw.csv')
```

```python
print("raw shape:", data.shape)
```

```python
data.tail(10)
```

```python
# 2) Basic cleanup: trim strings + convert common "null" text to NaN
obj_cols = data.select_dtypes(include=["object", "string"]).columns
data[obj_cols] = data[obj_cols].apply(lambda s: s.astype("string").str.strip())
data.replace({"": np.nan, "na": np.nan, "n/a": np.nan, "null": np.nan, "none": np.nan}, inplace=True)
```

```python
# 3) Drop fully-empty rows
data = data.dropna(how="all")
```

```python
# 4) Remove duplicates
data = data.drop_duplicates()
```

```python
print(data)
```

```python
# 5) Handle nulls: fill numeric with median, text with mode
num_cols = data.select_dtypes(include=[np.number]).columns
data[num_cols] = data[num_cols].apply(lambda s: s.fillna(s.median()))

obj_cols = data.select_dtypes(include=["object", "string"]).columns
for c in obj_cols:
    if data[c].isna().any():
        mode = data[c].mode(dropna=True)
        data[c] = data[c].fillna(mode.iloc[0] if len(mode) else "")
```
