```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('heart_disease_uci.csv')
```

```python
df.iloc[0:4]
```

```python
df['num'].unique()
```

```python
df.describe()
```

```python
##  Why this plot?

## Bar plots are ideal for comparing aggregated values across categories
## ere, we compare average heart disease severity (num) across Male vs Female

## Data transformation
## Used groupby("sex").mean() to aggregate disease severity

## Observed trend
## Males show higher average heart disease severity than females
## This aligns with known clinical risk patterns
```

```python
df.groupby("sex")["num"].mean().plot(
    kind="bar",
    title="Average Heart Disease Severity by Sex",
    ylabel="Average Disease Level",
    xlabel="Sex",
    color="steelblue"
)
plt.show()
```

```python
##Why this plot?

##Line plots show trends over ordered variables

##Age is continuous and ordered, making it ideal for trend analysis

##Data transformation

##Grouped by age and calculated mean disease severity

##Observed trend

##Heart disease severity increases with age

##Sharp rise after age ~50, indicating higher risk in older populations
```

```python
df.groupby("age")["num"].mean().plot(
    kind="line",
    title="Heart Disease Severity vs Age",
    ylabel="Average Disease Level",
    xlabel="Age"
)
plt.show()
```

```python
## Why this plot?

## Histograms are best for understanding data distribution

## Cholesterol is a continuous medical measurement

## Data transformation

## None (raw distribution)

## Observed trend

## Most patients fall between 200–300 mg/dL
## Right-skewed distribution indicates some patients with very high cholesterol
```

```python
df["chol"].plot(
    kind="hist",
    bins=30,
    title="Distribution of Cholesterol Levels",
    xlabel="Cholesterol",
    color="orange"
)
plt.show()
```

```python
## Why this plot?

## Scatter plots reveal relationships between two continuous variables

## Useful for spotting correlation or clusters

## Data transformation

## None required

## Observed trend

## Weak negative relationship:

## Higher cholesterol often corresponds to lower maximum heart rate

## Suggests reduced cardiovascular performance in high-cholesterol patients
```

```python
df.plot(
    kind="scatter",
    x="chol",
    y="thalch",
    title="Cholesterol vs Maximum Heart Rate",
    alpha=0.6
)
plt.show()
```
